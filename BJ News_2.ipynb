```python
import mechanicalsoup
import pandas as pd

# Initialize browser
browser = mechanicalsoup.StatefulBrowser()
browser.open("https://www.bjnews.com.cn/estate")

# Parse page content
page = browser.page
article_elements = page.select('div.col-xs-12.col-lg-4.hotlist')

# Extract titles and links
articles = []
for article in article_elements:
    link = article.find('a', href=True)
    title_div = link.find("div", class_="pin_tit") if link else None
    title = title_div.get_text(strip=True) if title_div else None
    url = link['href'] if link else None
    
    if title and url:
        articles.append({'Title': title, 'URL': url})

# Convert to DataFrame for easy manipulation and save to CSV
df = pd.DataFrame(articles)
df.to_csv("BJNews_Articles.csv", index=False)
```

```python
import mechanicalsoup
import pandas as pd

# Initialize the browser
browser = mechanicalsoup.StatefulBrowser()

# Base URL for pagination
base_url = "https://www.bjnews.com.cn/estate"

# Number of pages to scrape
max_pages = 20

# Storage for article data
articles = []

# Loop through the specified number of pages
for page_num in range(1, max_pages + 1):
    # Construct the page URL
    if page_num == 1:
        url = base_url  # First page
    else:
        url = f"{base_url}/{page_num}.html"  # Subsequent pages
    
    # Open the page
    browser.open(url)
    print(f"Scraping page {page_num}: {url}")
    
    # Get the page content and parse the articles
    page = browser.page
    article_elements = page.select('div.col-xs-12.col-lg-4.hotlist')  # Adjust this selector if needed

    # Debug: Check if articles are being found
    print(f"Found {len(article_elements)} articles on page {page_num}.")

    for article in article_elements:
        # Extract link and title
        link = article.find('a', href=True)
        title_div = link.find("div", class_="pin_tit") if link else None
        title = title_div.get_text(strip=True) if title_div else None
        url = link['href'] if link else None

        # Debug: Print title and URL to verify extraction
        print(f"Title: {title}, URL: {url}")

        # Store article data if both title and URL are found
        if title and url:
            articles.append({'Title': title, 'URL': url})

# Convert to DataFrame and save to CSV
df = pd.DataFrame(articles)
df.to_csv("BJNews_First_10_Pages.csv", index=False)

```

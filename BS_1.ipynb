```python
# Import package 
from bs4 import BeautifulSoup
```

```python
# Read a sample html file 
with open('home.html', 'r') as html_file: 
    content = html_file.read()
    print(content)
    
```

```python
# See the html code in a more preety way, Method: lxml 
soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

```python
# Grad information based on the tag
tags = soup.find_all('a')
print(tags)

tags = soup.find_all('a')
for i in tags: 
    print(i.text)
```

```python
soup = BeautifulSoup(content, 'lxml')
section = soup.find_all('ul', class_='jcul')

for title in section: 
    print(title)
```

```python
# Print text and link from above 
for title in section: 
    title_names = title.text
    title_links = title.li
    
print(title_names)
print(title_links)
```

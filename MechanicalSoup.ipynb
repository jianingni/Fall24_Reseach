```python
import mechanicalsoup
```

```python
# Open the browser
browser = mechanicalsoup.StatefulBrowser()
browser.open("https://www.bjnews.com.cn/estate")
```

```python
# Get the page content
page = browser.page

# Select the main content
content = page.select_one('div.row')
content_list = content.select_one('div.col-xs-12.col-lg-12')
article_list = content_list.select('div.col-xs-12.col-lg-4.hotlist') #this is a list of articles 

content_list
article_list

# Loop through the articles to extract both link and title
for article in article_list: 
    info_full = article.find_all("a")

    for info in info_full:
        # Extract the news link
        news_link = info.get("href")
        if news_link:
            print(news_link)

        # Extract the news title from div with class pin_tit
        news_title = info.select('div.pin_tit')
        if  news_title:
            news_title_1 = news_title.get_text().strip()
            print(news_title_1)
```
